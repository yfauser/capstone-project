```python
import pandas as pd
import pyarrow.parquet as pq
from visualization import save_specto
from sklearn.model_selection import train_test_split
```

```python
train_data = pq.read_pandas('data/select_train.parquet').to_pandas()
train_meta = pd.read_csv('data/metadata_train.csv')
```

```python
train_data.head()
```

```python
train_pd_positive = 'Data/train_spectograms/pd_positive'
train_pd_negative = 'Data/train_spectograms/pd_negative'
valid_pd_positive = 'Data/valid_spectograms/pd_positive'
valid_pd_negative = 'Data/valid_spectograms/pd_negative'
```

```python
train_meta_pd = train_meta[train_meta['target'] == 1]
```

```python
train_meta_pd.head()
```

```python
test_percentage = 20
```

```python
test_ratio = 100/test_percentage
images_total = train_data.shape[1]
for idx, column in enumerate(train_data.columns):
    images_left = images_total - idx
    
    if int(column) in train_meta_pd.index:
        if images_left % test_ratio:
            filepath = train_pd_positive
        else:
            filepath = valid_pd_positive
    else:
        if images_left % test_ratio:
            filepath = train_pd_negative
        else:
            filepath = valid_pd_negative
        
    print('working on image {}.png, {} images left to process'.format(column, images_left), end="\r")
    save_specto(train_data[column], filepath=filepath, filename='{}.png'.format(column))
```
